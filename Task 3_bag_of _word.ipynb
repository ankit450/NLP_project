```python
import string
from collections import Counter
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
# from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import warnings
warnings.filterwarnings('ignore')
```

```python
# for bag of words
# let's import neessary packages

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
# let's load a first document
# reading text file

text = open('National_news.txt' , encoding="utf-8").read()

print(text)
```

```python
# first step it will convert in lower case of all pragraph

Lower_document = text.lower()
print(Lower_document)
```

```python
# let's clear the documents by removing spaces and punctuation
cleaned_text = Lower_document.translate(str.maketrans('','', string.punctuation))
```

```python
cleaned_text
```

```python
# let's splitting text into words

tokenization_words = word_tokenize(cleaned_text, 'english')
```

```python
tokenization_words
print(len(tokenization_words))
```

```python
# removing word from the tockenization words list
final_words =[]
for word in tokenization_words:
    if word not in stopwords.words('english'):
        final_words.append(word)
        
```

```python
# Lemmatization from plural to single + Base from of a word 

lemma_word=[]

for word in final_words:
    word= WordNetLemmatizer().lemmatize(word)
    lemma_word.append(word)
```

```python
lemma_word
print(len(lemma_word))
```

```python
# let's call vectorization method

vectore1 = CountVectorizer()
print(vectore1)
```

```python
# this step will convert text into tokens 
vectore1.fit_transform(lemma_word)
print('Bag of words :',vectore1.get_feature_names())
```

```python
vectore1.vocabulary_
```

```python
lemma_word1=[]

for word in cleaned_text:
    word= WordNetLemmatizer().lemmatize(word)
    lemma_word1.append(word)
```

```python
# this step will convert text into tokens 
vector.fit_transform(lemma_word1)
print('Bag of words :',vector.get_feature_names())
```
